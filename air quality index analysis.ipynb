```python
import numpy as pd
import pandas as pd
```

```python
# loading the data from csv file to pandas dataframe
air_quality_data = pd.read_csv('/content/data_date.csv')
```

```python
# printing the first 10 rows of the dataframe
air_quality_data.head()
```

```python
# @title Status

from matplotlib import pyplot as plt
import seaborn as sns
air_quality_data.groupby('Status').size().plot(kind='barh', color=sns.palettes.mpl_palette('Dark2'))
plt.gca().spines[['top', 'right',]].set_visible(False)
```

```python
# @title AQI Value

from matplotlib import pyplot as plt
air_quality_data['AQI Value'].plot(kind='hist', bins=20, title='AQI Value')
plt.gca().spines[['top', 'right',]].set_visible(False)
```
